```python
import pandas as pd
import matplotlib.pyplot as plt

# Input Data
data = {
    "Year": [1810, 1820, 1830, 1840, 1850, 1860, 1870, 1880, 1900, 1910, 1920, 1930, 1940, 1950, 1960, 1970, 1980],
    "Height_cm": [169.7, 169.1, 166.7, 166.5, 165.6, 166.6, 167.2, 168, 169.4, 170.9, 171, 173.9, 174.9, 176, 176.9, 177.1, 176.8]
}

# Create DataFrame
df = pd.DataFrame(data)

# Calculate Mean Height
mean_height = df["Height_cm"].mean()
print(f"Mean Height from 1810 to 1980: {mean_height:.2f} cm")

# Visualization
plt.figure(figsize=(10, 6))
plt.plot(df["Year"], df["Height_cm"], marker='o', linestyle='-', linewidth=2)
plt.title("Average Male Height in UK (1810 - 1980)")
plt.xlabel("Year")
plt.ylabel("Height (cm)")
plt.grid(True)
plt.tight_layout()
plt.show()
```
